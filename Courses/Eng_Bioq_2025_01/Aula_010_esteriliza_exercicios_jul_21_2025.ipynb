Discuta qual a temperatura associada a curva C e qual o valor previsto de N/N0 sobrevida associado a um tratamento com injeção de vapor e tempo de residência no tubo de espera de 40s a 148$^0$C.

```python
# curva A: t = 10 min tem N = 10^2 e t=0 tem N=10^6--> k*tempo = ln(N0/N)--> k1 = 1/tempo*ln(N0/N)
# curva C: t = 55 min tem N = 1   e t= 0 tem N= 10^6 -> k2 = tempo*ln(N0/N)
tempo = 10
N0_N = 10**6/10**2
k1 = 1/tempo*ln(N0_N)
tempo = 55
N0_N = 10**6/10**0
k2 = 1/ tempo*ln(N0_N)
print('k1=',k1,'k2=',k2)
```

```python
T1 = 60+273.15
T2 = 55+273.15
R = 8.314
Ea = -R*ln(k1/k2)/(1/T1-1/T2)
print('Ea=',Ea, 'J/mol')
```

```python
# curva B: t=30 N=10**0 e t=0 N=10**6
tempo = 30
N0_N = 10**6/10**0
k3 = 1/tempo*ln(N0_N)
print('k3=',k3, 'min-1')
```

```python
inv_T3 = -R*ln(k3/k2)/Ea+1/T2
T3 = 1/inv_T3 - 273.15
print('T3=',T3)
```

```python
T = 148+ 273.15
theta = 40/60
k = k1*exp(-Ea/R*(1/T-1/T1))
N_N0 = exp(-k*theta)
print(f'N/N0={N_N0:.2e}')
```
