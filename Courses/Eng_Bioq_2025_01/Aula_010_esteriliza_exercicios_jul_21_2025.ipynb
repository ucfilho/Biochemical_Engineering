**Questao 3** A produção do produto WZ segue a seqüência ininterrupta de fermentação, filtração,envaze e esterilização a 100$^0$C com jatos de água e vapor. Pelas características da embalagem a passagem da mesma diretamente para 100$^0$C é indesejável por gerar a deformação. Assim a esterilização é feita com a condução do material embalado por um túnel que passa pelas regiões de temperatura de 40, 60, 100 $^0$C conforme mostra a figura. Nestas condições pede-se a)considerando desprezível a morte celular nas regiões e jato de água, calcule a velocidade constante que o vagão deve ser mover para que a esterilização atinja uma sobrevida de 10$^{-6}$. b) para o caso de morte não desprezível nas regiões de jato de água quente, calcule a velocidade constante que o vagão deve se mover para que a esterilização atinja a mesma sobrevida do item (a). Obs:considere que a temperatura do interior da embalagem se torna a mesma de cada região após 3 min que o produto se encontra na região ,equação de morte celular com k expresso em min$^{-1}$ e T em kelvin: </br>
ko=1,2x10^38 e Ea=68,7 kcal/mol.K com k=ko*exp(−Ea/RT)

image.png

```python
# Area = (k0+k1)/2*3+(k1+k2)/2*3+(k2+k3)/2*3+ (k1*L1+k2*L2+k3*L3)/u-3*(k1+k2+k3)
# Area = ln(N0/N)
A, Ea, R =1.2e38, 68.7e3, 1.987
L1, L2, L3 = 10,10,25
N_N0 =10**(-6)
N0_N = 1/N_N0
Area = ln(N0_N)
T0 = 20+273.15; T1 = 40+273.15;T2 = 60+273.15;T3 = 100+273.15
k0, k1= A*exp(-Ea/(R*T0)),A*exp(-Ea/(R*T1));
k2, k3 = A*exp(-Ea/(R*T2)), A*exp(-Ea/(R*T3));
parte1 = (k0+k1)/2*3+(k1+k2)/2*3+(k2+k3)/2*3
parte3 = 3*(k1+k2+k3)
#parte2 = (k1*L1+k2*L2+k3*L3)/u
#Area = parte1 + parte2 - parte3
parte2 = Area -parte1+ parte3
u = (k1*L1+k2*L2+k3*L3)/parte2
print('u=',u, 'm/min')
```

```python
Mi=0.2;# viscosidade cPoise
Ro=1020;# densidade kg/m3
Mi=Mi/1000;


Vol= 100e3 #"entre com volume (litros)-->"
Tempo_processo=6 #"entre com tempo (horas) de processo (dado operacional)-->"

Tempo_processo=Tempo_processo*3600




# set up the Axes for the first plot
N_N0 = 1e-6
TC = 130
k=3.73e-4*exp(0.0858875*TC);
k=k/60;
Vol = Vol/1000
# plot a 3D surface like in the example mplot3d/surface3d_demo
U_set= np.linspace(10, 60, num=40); n = U_set.shape[0]
d_set= np.linspace(15, 30, num=40); m = d_set.shape[0]
Num = np.zeros((n, m))
L_set = np.zeros((n, m))
for i in range(n):
    for j in range(m):
      d=d_set[j]/100;U=U_set[i]/100;
      Area=pi*d**2/4;
      Vazao=U*Area;
      Re=Ro*U*d/Mi;
      Theta=1/k*ln(1/N_N0);
      L=U*Theta;
      Vazao=U*Area;
      L_set[j, i] = L
      Num_tubos = ceil(Vol/(Vazao*Tempo_processo));

      Num[j,i ]= Num_tubos
# =============
# First subplot
# =============
# set up a figure twice as wide as it is tall
#fig = plt.figure(figsize=plt.figaspect(0.5))
fig = plt.figure(figsize=(12, 4))
ax = fig.add_subplot(1, 2, 1, projection='3d')
U, d = np.meshgrid(U_set, d_set)

surf = ax.plot_surface(U, d, Num, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
# Set the axis labels
ax.set_xlabel('U-cm/s')
ax.set_ylabel('d-cm')
ax.set_zlabel('Equipamentos')
#fig.colorbar(surf, shrink=0.5, aspect=10)

# ==============
# Second subplot
# ==============
# set up the Axes for the second plot
ax = fig.add_subplot(1, 2, 2, projection='3d')

surf2 = ax.plot_surface(U, d, L_set, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.view_init(elev=30, azim=70) # Example: 30 degrees elevation, 45 degrees azimuth
# Set the axis labels
ax.set_xlabel('U-cm/s')
ax.set_ylabel('d-cm')
ax.set_zlabel('L-metros')
fig.subplots_adjust(wspace=0.3)
#plt.tight_layout()
plt.show()
```
