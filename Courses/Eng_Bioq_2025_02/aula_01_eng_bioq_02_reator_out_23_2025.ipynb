```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
Yxs = 0.5
Ypx  = 0.2
S0= 100
P0= 0
X0 = 0
F0 = 100
F = F0
Vol = 500
mi_m = 0.4
Ks = 2
D = F/Vol
#dX_dt = F0/Vol*X0-F/Vol*X+mi*X
#dS_dt = F0/Vol*S0-F/Vol*S-mi*X/Yxs
#dP_dt = F0/Vol*P0-F/Vol*P+mi*X*Ypx
```

```python
y0, t0 = [100.0, 10.0, 10.0], 0
```

```python
t = np.linspace(0, 500, 101)
```

```python
def f(y, t):
    X,S,P = y
    mi= mi_m*S/(S+Ks)
    dX_dt = F0/Vol*X0-F/Vol*X+mi*X
    dS_dt = F0/Vol*S0-F/Vol*S-mi*X/Yxs
    dP_dt = F0/Vol*P0-F/Vol*P+mi*X*Ypx
    return [dX_dt,dS_dt,dP_dt]
```

```python
sol = odeint(f, y0, t)
```

```python
plt.plot(t, sol[:, 1], 'b', label='S(t)')
```

```python
plt.plot(t, sol[:, 0], 'r', label='S(t)')
```

```python
plt.plot(t,sol)
plt.legend(['X(t)','S(t)','P(t)'],loc="upper right")
plt.xlabel('tempo-h')
plt.ylabel('Conc- g/L')
```
