```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('calcula_motor.xlsx')
```

```python
df
```

```python
nrows, ncols = df.shape
print(nrows, ncols)
```

```python
col_names  = ['Vol', 'Nrps', 'Np', 'Nrps', 'ro', 'mi', 'd', 'di', 'Re', 'Pot']
data = pd.DataFrame(columns=col_names)
for i in range(nrows):
  Vol = df.loc[i, 'Vol']
  Nrps = df.loc[i, 'Nrps']
  Np = df.loc[i, 'Np']
  Nrps = df.loc[i, 'Nrps']
  ro = df.loc[i, 'ro']
  mi = df.loc[i, 'mi']
  d = (4*Vol/np.pi)**(1/3)
  di = d/3
  Pot = Np*ro*di**5*Nrps**3
  Re = ro*di**2*Nrps/mi
  resp = {}
  resp['Vol'] = Vol
  resp['Nrps'] = Nrps
  resp['Np'] = Np
  resp['Nrps'] = Nrps
  resp['ro'] = ro
  resp['mi'] = mi
  resp['d'] = d
  resp['di'] = di
  resp['Re']=Re
  resp['Pot'] = Pot
  data.loc[i] = resp
```

```python
data
```

```python
data.to_excel('motor.xlsx')
```
